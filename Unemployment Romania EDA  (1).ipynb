```python
df2_medii.pivot_table(
    ["NUMAR TOTAL SOMERI FEMEI  ", "NUMAR TOTAL SOMERI BARBATI "],
    ["JUDET "],
    aggfunc = ["sum"]
)
```

```python
X = df2_medii[['NUMAR TOTAL SOMERI ']]
y = df2_medii[['NUMAR TOTAL SOMERI DIN MEDIUL RURAL ']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_test, y_test)
```

```python
y_pred=regressor.predict(X_test)
evaluate = pd.DataFrame({'Actual': y_test.values.flatten(), 'Predicted': y_pred.flatten()})
evaluate.head(10)
```
